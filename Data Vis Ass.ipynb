# **Ans 5:-**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(59)

n_samples = 100
x1 = np.random.normal(loc=0, scale=1, size=n_samples)
x2 = x1 + np.random.normal(loc=0, scale=0.5, size=n_samples)
x3 = np.random.normal(loc=0, scale=1, size=n_samples)
x4 = 0.5 * x2 + 0.3 * x3 + np.random.normal(loc=0, scale=0.5, size=n_samples)
```

```python
data = pd.DataFrame({'Feature 1': x1, 'Feature 2': x2, 'Feature 3': x3, 'Feature 4': x4})

correlation_matrix = data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})

plt.title('Correlation Matrix Heatmap')
plt.show()
```

# **Ans 3:-**

```python
# Generate sales data based on the provided instruction
np.random.seed(20)
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 32), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data_sales)
```

```python
# Pivot the data for heatmap visualization (aggregating by the mean sales for each month and day)
sales_pivot = df_sales.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(sales_pivot, cmap="YlGnBu", annot=True, fmt=".0f")

plt.title('Heatmap of Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

# **Ans 4**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate z = sin(sqrt(x^2 + y^2))
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create the 3D surface plot
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Surface Plot: z = sin(sqrt(x^2 + y^2))')

# Add a color bar
plt.colorbar(surf, ax=ax, shrink=0.7, aspect=7)

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#  data
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100000, 1000000, 5),
    'GDP': np.random.randint(500000, 5000000, 5)
}

df = pd.DataFrame(data)

# Bubble chart
plt.figure(figsize=(10, 6))

# bubble sizes based on population
plt.scatter(df['GDP'], df['Population'], s=df['Population'] / 1000, alpha=0.5, c='b')

# Adding each country
for i in range(df.shape[0]):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=12)

plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: Population vs GDP')

plt.show()
```

# **Ans 3:-**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral11

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

output_file("fruit_counts.html")

p = figure(x_range=fruits, height=400, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9, line_color='white',
       fill_color=factor_cmap('x', palette=Spectral11, factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
p.xaxis.major_label_orientation = "vertical"

show(p)
```

# **Ans 5:-**

```python
import numpy as np
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palette import Viridis256

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title="Heatmap Example", x_axis_label='X-axis', y_axis_label='Y-axis',
           x_range=(0, 1), y_range=(0, 1),
           tools="", toolbar_location=None)

mapper = linear_cmap('value', Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
